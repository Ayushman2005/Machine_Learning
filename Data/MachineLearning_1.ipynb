```python
import json
```

```python
#Load the data
def load_data(filename):
  with open(filename,"r") as f:
    data = json.load(f)
  return data
```

```python
data = load_data("store_data.json")
print(data, type(data))
```

```python
def clean_data(data):
  text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
  cleaned_data=[]
  unique_users = set()
  for user in data:
    raw_rating = str(user["rating"]).strip().lower()
    if(raw_rating in text_to_num):
      raw_rating = text_to_num[raw_rating]
    user["rating"] = raw_rating
    raw_age = user.get("age")
    if(raw_age == None):
      user["age"] = None
    if(user["name"].strip() in unique_users):
      continue
    unique_users.add(user["name"])
    cleaned_data.append(user)
  return cleaned_data
```

```python
clean_data(data)
```

```python
def get_insights(data):
  tot_rating = 0
  for user in data:
    tot_rating += float(user["rating"])
  print(f"avg rating = {tot_rating/len(data)}")
  poor_ratings = 0
  for user in data:
    if(float(user["rating"])<3):
      poor_ratings += 1
  print(f"% users with poor rating = {poor_ratings/len(data)*100}%")
```

```python
get_insights(data)
```
